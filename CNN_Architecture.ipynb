1. What is the role of filters and feature maps in Convolutional Neural
Network (CNN)?
   - In a Convolutional Neural Network (CNN), filters and feature maps play a central role in helping the model understand and extract meaningful patterns from data, especially images. Filters, also known as kernels, are small matrices of weights that slide over the input (like an image) and look for specific patterns such as edges, textures, or shapes. Each filter is designed or learned during training to detect a particular feature for example, one filter might highlight vertical edges while another captures curves or colors. As a filter moves across the input, it performs a mathematical operation convolution at each position, producing a new set of values. The result of applying a filter is called a feature map. This feature map shows where and how strongly a particular feature is present in different parts of the input. In simpler terms, while filters act like “pattern detectors,” feature maps are the “highlighted results” showing where those patterns exist. Together, multiple filters generate multiple feature maps, allowing the CNN to capture a wide variety of features from the input. As the network goes deeper, these feature maps become more abstract, combining simple features into more complex representations, which ultimately helps the model make accurate predictions.

2. Explain the concepts of padding and stride in CNNs(Convolutional Neural
Network). How do they affect the output dimensions of feature maps?
   - In Convolutional Neural Networks (CNNs), padding and stride are two important concepts that control how filters move over the input and how the size of the resulting feature maps is determined. Padding refers to adding extra pixels usually zeros around the border of the input image before applying the convolution. This is done to control the spatial size of the output. Without padding called valid padding, the feature map becomes smaller because the filter cannot go beyond the edges of the image. With padding often called same padding, extra pixels are added so that the output feature map can retain the same height and width as the input. Padding is especially useful when we want to preserve edge information and avoid excessive shrinking of the data as it passes through multiple layers. Stride, on the other hand, defines how much the filter moves or “steps” across the input image. A stride of 1 means the filter moves one pixel at a time, covering every possible position and producing a larger feature map. A stride of 2 or more means the filter jumps multiple pixels at once, which reduces the size of the output feature map because fewer positions are evaluated. In simple terms, a larger stride leads to a smaller output, while a smaller stride keeps more detailed spatial information. Together, padding and stride directly affect the output dimensions of feature maps. Padding tends to increase or preserve the size, while stride reduces it. By carefully choosing these parameters, we can control how much spatial detail is retained and how computationally efficient the network becomes.
3. Define receptive field in the context of CNNs. Why is it important for deep architectures?
   - In Convolutional Neural Networks (CNNs), the receptive field refers to the portion of the input image that influences a particular neuron in a feature map. In simple terms, it is the area of the original image that a neuron “looks at” to produce its output. In the initial layers, the receptive field is small and limited to local regions based on filter size, but as we move deeper into the network, it increases because each layer builds on the outputs of previous layers. The concept is important for deep architectures because it allows the network to capture both local and global information. Early layers detect simple features like edges and textures, while deeper layers, with larger receptive fields, can recognize more complex patterns such as shapes or entire objects. This gradual expansion helps the model understand the overall context of the image. A well-designed receptive field ensures that the network does not miss important spatial relationships in the input. If it is too small, the model may fail to capture larger patterns, if sufficiently large, it improves the model’s ability to make accurate predictions by considering broader context.

4. Discuss how filter size and stride influence the number of parameters in a CNN.
   - In a Convolutional Neural Network (CNN), both filter size and stride play different roles when it comes to the number of parameters in the model. The filter size has a direct impact on the number of parameters. Each filter contains weights that are learned during training, and the total number of parameters depends on its dimensions. For example, a larger filter like 5×5 will have more weights than a smaller one like 3×3, and therefore increases the total number of parameters in the network. If there are multiple filters in a layer, the parameter count increases further. So, using larger filters makes the model more complex and computationally expensive, while smaller filters help keep the parameter count lower and are commonly preferred in modern CNN architectures. On the other hand, stride does not directly affect the number of parameters. Stride only controls how the filter moves across the input the step size. Whether the stride is 1 or 2, the filter itself and thus its weights remains the same, so the number of parameters does not change. However, stride does affect the size of the output feature map. A larger stride reduces the spatial dimensions of the output, which can indirectly reduce computational cost in later
   layers, but it does not change the parameter count of the convolutional layer itself.

5. Compare and contrast different CNN-based architectures like LeNet,
AlexNet, and VGG in terms of depth, filter sizes, and performance.
   - Convolutional Neural Networks (CNNs) have evolved significantly over time, and architectures like LeNet, AlexNet, and VGG highlight this progression in terms of depth, filter sizes, and performance. LeNet, one of the earliest CNN architectures developed by Yann LeCun, is relatively shallow with about 5–7 layers. It uses small filter sizes mostly 5×5 and was designed mainly for simple tasks like handwritten digit recognition. Due to its simplicity, LeNet has a low number of parameters and works well on smaller datasets, but its performance is limited for complex image recognition tasks. AlexNet, introduced by Alex Krizhevsky and his team, marked a major breakthrough in deep learning. It is deeper than LeNet, consisting of 8 layers 5 convolutional and 3 fully connected layers. It uses a mix of filter sizes, including large filters like 11×11 in the initial layers, followed by smaller ones. AlexNet significantly improved performance on large scale image datasets like ImageNet due to innovations such as ReLU activation, dropout, and GPU training. However, it has a relatively high number of parameters.VGG, developed by the Visual Geometry Group, pushed depth even further, with configurations like VGG-16 and VGG-19 containing 16 and 19 layers respectively. Unlike AlexNet, VGG uses very small filters 3×3 consistently throughout the network. This design increases depth while keeping the architecture uniform and easier to manage. VGG achieves higher accuracy than earlier models, but at the cost of a very large number of parameters and high computational requirements.In comparison, LeNet is shallow and simple with moderate filter sizes, AlexNet is moderately deep with varied filter sizes and improved performance, while VGG is very deep with small filters and achieves better accuracy but is computationally expensive. Overall, as we move from LeNet to VGG, the depth increases, filter sizes become smaller and more consistent, and performance improves significantly on complex image recognition tasks.



```python
# Using keras, build and train a simple CNN model on the MNIST dataset
# from scratch. Include code for module creation, compilation, training, and evaluation.
# (Include your Python code and output in the code box below.)

# Import required libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize and reshape data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Reshape to include channel dimension (28x28x1)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Build the CNN model
model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Display model summary
model.summary()

# Train the model
history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.1
)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(x_test, y_test)

print("\nTest Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
# Load and preprocess the CIFAR-10 dataset using Keras, and create a
# CNN model to classify RGB images. Show your preprocessing and architecture.
# (Include your Python code and output in the code box below.)

# Import libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

# Preprocessing
# Normalize pixel values (0–255 → 0–1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# Convert labels to categorical (one-hot encoding)
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# Check shapes
print("Training data shape:", x_train.shape)
print("Test data shape:", x_test.shape)

# Build CNN model
model = keras.Sequential([
    # First Convolution Block
    layers.Conv2D(32, (3,3), activation='relu', padding='same', input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    # Second Convolution Block
    layers.Conv2D(64, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),

    # Third Convolution Block
    layers.Conv2D(128, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),

    # Fully Connected Layers
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Model summary
model.summary()

# Train model
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1
)

# Evaluate model
test_loss, test_acc = model.evaluate(x_test, y_test)

print("\nTest Loss:", test_loss)
print("Test Accuracy:", test_acc)
```

```python
# Using PyTorch, write a script to define and train a CNN on the MNIST
# dataset. Include model definition, data loaders, training loop, and accuracy evaluation.
# (Include your Python code and output in the code box below.)

# Import libraries
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load and preprocess MNIST dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_dataset = torchvision.datasets.MNIST(
    root='./data', train=True, transform=transform, download=True)

test_dataset = torchvision.datasets.MNIST(
    root='./data', train=False, transform=transform, download=True)

train_loader = torch.utils.data.DataLoader(
    dataset=train_dataset, batch_size=64, shuffle=True)

test_loader = torch.utils.data.DataLoader(
    dataset=test_dataset, batch_size=64, shuffle=False)

# Define CNN model
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)

        self.fc1 = nn.Linear(64 * 5 * 5, 128)
        self.fc2 = nn.Linear(128, 10)

        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))  # (28→26→13)
        x = self.pool(self.relu(self.conv2(x)))  # (13→11→5)
        x = x.view(-1, 64 * 5 * 5)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = CNN().to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 5

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}")

# Evaluation
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)

        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f"\nTest Accuracy: {accuracy:.2f}%")
```

```python

# Given a custom image dataset stored in a local directory, write code using
# Keras ImageDataGenerator to preprocess and train a CNN model.
# (Include your Python code and output in the code box below.)

# Import libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models

# Set dataset paths
train_dir = 'dataset/train'
val_dir = 'dataset/validation'

# Image preprocessing using ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    shear_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

# Load images from directories
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

# Build CNN model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Model summary
model.summary()

# Train model
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator
)

# Evaluate model
loss, accuracy = model.evaluate(val_generator)

print("\nValidation Loss:", loss)
print("Validation Accuracy:", accuracy)
```

10. You are working on a web application for a medical imaging startup. Your
task is to build and deploy a CNN model that classifies chest X-ray images into “Normal”
and “Pneumonia” categories. Describe your end-to-end approach–from data preparation
and model training to deploying the model as a web app using Streamlit.
(Include your Python code and output in the code box below.)
 - The end-to-end approach to building a CNN-based web application for classifying chest X-ray images into “Normal” and “Pneumonia” involves four main stages: data preparation, model training, evaluation, and deployment First, in data preparation, the dataset is organized into training and validation folders with separate classes. Images are resized, normalized, and augmented (e.g., rotation, flipping) to improve model performance and generalization. Next, in model training, a CNN is built using convolutional, pooling, and fully connected layers. The model is trained using an optimizer like Adam and a loss function such as binary cross-entropy. It learns to identify patterns that distinguish normal lungs from pneumonia cases. After training, evaluation is performed on validation or test data to measure accuracy and ensure the model performs well on unseen images. Finally, in deployment, the trained model is integrated into a Streamlit web app. Users can upload X-ray images, which are preprocessed and passed to the model for prediction. The app then displays whether the image is Normal or Pneumonia along with confidence.

```python


import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import streamlit as st
from PIL import Image

# PATHS
TRAIN_DIR = "dataset/train"
VAL_DIR = "dataset/val"
MODEL_PATH = "xray_model.h5"


# TRAIN MODEL FUNCTION
def train_model():
    train_gen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        zoom_range=0.2,
        horizontal_flip=True
    )

    train_data = train_gen.flow_from_directory(
        TRAIN_DIR,
        target_size=(224,224),
        batch_size=32,
        class_mode='binary'
    )

    val_data = ImageDataGenerator(rescale=1./255).flow_from_directory(
        VAL_DIR,
        target_size=(224,224),
        batch_size=32,
        class_mode='binary'
    )

    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
        layers.MaxPooling2D(2,2),

        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D(2,2),

        layers.Conv2D(128, (3,3), activation='relu'),
        layers.MaxPooling2D(2,2),

        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(train_data, epochs=5, validation_data=val_data)

    model.save(MODEL_PATH)
    return model

# LOAD OR TRAIN MODEL
if os.path.exists(MODEL_PATH):
    model = tf.keras.models.load_model(MODEL_PATH)
else:
    model = train_model()

# STREAMLIT APP
st.title("Chest X-ray Classifier (Normal vs Pneumonia)")
st.write("Upload an X-ray image to get prediction")

uploaded_file = st.file_uploader("Choose an image", type=["jpg","png","jpeg"])

def preprocess_image(image):
    image = image.resize((224,224))
    image = np.array(image) / 255.0
    image = np.expand_dims(image, axis=0)
    return image

if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption="Uploaded Image", use_column_width=True)

    img = preprocess_image(image)
    prediction = model.predict(img)

    if prediction[0][0] > 0.5:
        result = "Pneumonia"
    else:
        result = "Normal"

    st.subheader(f"Prediction: {result}")
    st.write(f"Confidence: {float(prediction[0][0]):.2f}")

```
